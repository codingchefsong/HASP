```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
files = ["20230101_devttyUSB2.csv","20230101_devttyUSB3.csv","20230101_devttyUSB4.csv","20230101_devttyUSB6.csv"]
for file in files:
    df = pd.read_csv(file, sep=',',header=None,on_bad_lines="skip")
    df.head()
    # df.describe()
    
    # DataFrame
    # df[2] = pd.to_datetime(df[2], format='%H:%M:%S.%f').dt.time  # Assuming column 2 contains the x-axis values
    
    # temp
    df[8] = pd.to_numeric(df[8], errors='coerce') 
    df[24] = pd.to_numeric(df[24], errors='coerce')  
    df[26] = pd.to_numeric(df[26], errors='coerce')     
    # mag
    df[16] = pd.to_numeric(df[16], errors='coerce')
    df[18] = pd.to_numeric(df[18], errors='coerce')
    df[20] = pd.to_numeric(df[20], errors='coerce')
    # acc
    df[10] = pd.to_numeric(df[10], errors='coerce')
    df[12] = pd.to_numeric(df[12], errors='coerce')
    df[14] = pd.to_numeric(df[14], errors='coerce') 
    # cap
    df[22] = pd.to_numeric(df[22], errors='coerce')
    # timestamp
    df[6] = pd.to_numeric(df[6], errors='coerce') 
    
    
    # Create the plot
    fig, ((ax,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(3,2,figsize=(16, 16))
    
    # Plot the normalized datasets
    ax.plot(df[8],label='Temp')
    ax.plot(df[24],label='ObjTemp')
    ax.plot(df[26],label='AmbTemp')
    
    ax2.plot(df[10],label='accX')
    ax2.plot(df[12],label='accY')
    ax2.plot(df[14],label='accZ')
    
    ax3.plot(df[16],label='magX')
    ax3.plot(df[18],label='magY')
    ax3.plot(df[20],label='magZ')
    
    ax4.plot(df[22],label='cap')

    ax5.plot(df[6],label='Timestamp')

    

    
    # ax.set_ylim([-40, 80])
    ax.set_title(f"{df[3][0]} {df[4][0]} ({df[2][0]} - {df[2].iloc[-1]}) Temperature")
    ax.set_xlabel('Record') 
    ax.set_ylabel('Temperture(C)') 
    # ax.set_xticks(np.arange(0,len(df[2]),30*30))
    # ax.set_yticks(np.arange(-100,200,20))
    ax.legend()
    ax.grid(True)
    
    ax2.set_title(f"{df[3][0]} {df[4][0]} ({df[2][0]} - {df[2].iloc[-1]}) Accelerometer")
    ax2.set_xlabel('Record') 
    ax2.set_ylabel('Accelerotion') 
    ax2.legend()
    ax2.grid(True)
    
    ax3.set_title(f"{df[3][0]} {df[4][0]} ({df[2][0]} - {df[2].iloc[-1]}) Magnetometer")
    ax3.set_xlabel('Record') 
    ax3.set_ylabel('Magnetotion') 
    ax3.legend()
    ax3.grid(True)
    
    ax4.set_title(f"{df[3][0]} {df[4][0]} ({df[2][0]} - {df[2].iloc[-1]}) Capacitive")
    ax4.set_xlabel('Record') 
    ax4.set_ylabel('Magnetotion') 
    ax4.legend()
    ax4.grid(True)
    
    ax5.set_title(f"{df[3][0]} {df[4][0]} ({df[2][0]} - {df[2].iloc[-1]}) Timestamp")
    ax5.set_xlabel('Record') 
    ax5.set_ylabel('Timestamp') 
    ax5.legend()
    ax5.grid(True)
    
    
    
    # using padding
    fig.tight_layout(pad=1.0)
    # Save file
    plt.savefig(f"{df[3][0]}_{df[4][0]}.png",dpi=100)
    # Display the plot
    plt.show()
    # print(df[8],df[24],df[26])
    
```
